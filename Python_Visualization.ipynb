```python
#Plot graph of no.(-100 to 100) and its square, using range function.
x = list(range(-100,101))
y = [i**2 for i in x] #list comprehension
plt.plot(x,y)
```

```python
from math import exp
x = list(range(0,30))
y = [exp(i) for i in x]
plt.plot(x,y)
```

```python
from math import log
x = list(range(1,101))
y = [log(i) for i in x]
plt.plot(x,y)
```

```python
#Plot multiple graphs at a time
from math import *
plt.figure(figsize = (16,9))
x = list(range(-10,11))
y = [i**2 for i in x]
plt.subplot(3,4,1) #3x4 grid, 1 = position of graph
plt.plot(x,y)
plt.title("Graph of Squares")

y = [i**3 for i in x]
plt.subplot(3,4,7)
plt.plot(x,y)
plt.title("Graph of Cubes")

y=[sin(i) for i in x]
plt.subplot(3,4,10)
plt.plot(x,y)
plt.title("Graph of sin(x)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips') #tips is an inbuilt dataset
#sns.get_dataset_names()
```

```python
tips.columns
```

```python
pd.crosstab(tips['smoker'],tips['sex'])
```

```python
pd.crosstab(tips['smoker'],tips['sex']).plot(kind='bar')
```

```python
#Strip Plot
sns.stripplot(y='tip', data=tips)#, jitter=False/True)
plt.show()
```

```python
sns.stripplot(x='sex',y='tip', data=tips)
plt.show()
```

```python
sns.stripplot(x='sex',y='tip', data=tips, jitter=True)
plt.show()
```

```python
sns.stripplot(x='sex',y='tip', data=tips, jitter=False)
plt.show()
```

```python
sns.stripplot(x='day',y='tip', data=tips, jitter=True)
plt.show()
```

```python
sns.stripplot(x='day',y='tip', data=tips, size=2, jitter=True)
plt.show()
#plt.ylabel=('tip ($)')
```

```python
#Swarm Plot
sns.swarmplot(x='day', y='tip', data=tips, size=6)
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='time')
plt.show()
```

```python
#Subplots
plt.subplot(2,2,1) #2x2 grid position of this plot 1
sns.boxplot(x='day', y='tip', data=tips)
```
